```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def boltzmann_factor(E, T, k=1.0):
    """
    Returns the unnormalized Boltzmann factor e^(-E / (k * T))
    for a given energy E and temperature T.
    """
    return np.exp(-E / (k * T))
```

```python
def plot_boltzmann_distribution():
    # Energy range
    E = np.linspace(0, 5, 200)

    # Different temperature values
    temperatures = [0.5, 1.0, 2.0, 5.0, 10.0]

    plt.figure(figsize=(8, 6))
    for T in temperatures:
        P = boltzmann_factor(E, T)
        # (Optional) Normalize the distribution if you want a proper PDF:
        # P /= np.trapz(P, E)

        plt.plot(E, P, label=f"T = {T}")

    plt.xlabel("Energy (E)")
    plt.ylabel("Probability (unnormalized)")
    plt.title("Boltzmann Factor for Different Temperatures")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_boltzmann_distribution()
```
